```python
import cv2
import os
import numpy as np
```

```python
img = cv2.imread('.\\sampleW.png', 0)
copy = img.copy()
cv2.imshow('sample',copy)
cv2.waitKey(0)
cv2.destroyAllWindows()

brightness = 100
contrast = 1000
copy = np.int16(copy)
copy = img * (contrast/127+1) - contrast + brightness
copy = np.clip(copy, 0, 255)
copy = np.uint8(copy)

cv2.imshow('copy',copy)
cv2.waitKey(0)
cv2.destroyAllWindows()

mser = cv2.MSER_create()

mser.setMinArea(100)
mser.setMaxArea(1000)

#

coordinates, bboxes = mser.detectRegions(copy)
for bbox in bboxes:
    x, y, w, h = bbox
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

regions = mser.detectRegions(img)            

hulls = [cv2.convexHull(p.reshape(-1, 1, 2)) for p in regions[0]]

cv2.polylines(copy, hulls, 1, (0, 255, 0))

mask = np.zeros((img.shape[0], img.shape[1], 1), dtype=np.uint8)
mask = cv2.dilate(mask, np.ones((150, 150), np.uint8))
for contour in hulls:
    cv2.drawContours(mask, [contour], -1, (255, 255, 255), -1)

    text_only = cv2.bitwise_and(img, img, mask=mask)
            
cv2.imshow('copy',copy)
cv2.waitKey(0)
cv2.imshow('mask',mask)
cv2.waitKey(0)
cv2.imshow('text',text_only)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
